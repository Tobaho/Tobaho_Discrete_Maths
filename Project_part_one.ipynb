**Subsets**

```python
def IsASubset(A,B):
  for a in A:
    if a not in B:
      return False
    return True
IsASubset(Amazingspiderman,U)
```

```python
def IsBSubset(B,C):
  for b in B:
    if b not in C:
      return False
    return True
IsBSubset(Nowayhome,U)
```

```python
def IsCSubset(C,A):
  for c in C:
    if c not in A:
      return False
    return True
IsCSubset(Farfromhome,U)
```

**Powersets**


```python
def DPpowerset(s):
  powerset= [[]]
  for i in s:
    for sett in powerset:
      powerset = powerset+ [sett + [i]]
  return powerset
```

```python
DPpowerset(Amazingspiderman)
```
