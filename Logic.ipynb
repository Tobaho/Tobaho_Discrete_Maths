I am going to create three propositions which I will describe below. The three proposoitions are three ingredients that are used to prepare a native dish called "Jollof rice". 

I am going to assign p, q, and r to represent these three required ingredients. The three ingredients are:

*   p= Tomatoes
*   q= Red bell pepper
*   r= Onions


```python
AddTomatoes = True 
AddRedbellpepper = True
AddOnions = False
```

##1b Compute the truth of the statement 

The first truth statement i will do is adding onions and adding tomatoes
 $$ r \land p$$

```python
AddOnions and AddTomatoes
```

The next statement is going to be adding onions and not adding red bell pepper $$ r \lor \lnot q$$

```python
AddOnions or not AddRedbellpepper
```

The next statment is "not adding onions or adding tomatoes" describe is $$ r \Rightarrow p$$

```python
not(AddOnions) or AddTomatoes
```

I am computing a truth table for this $$ \lnot (r \oplus p) ⇒ (p \lor q)$$ 

```python
truthtable = ttg.Truths(['p', 'q', 'r'], ['r xor p', '~(r xor p)', 'p or q', '(~(r xor p)) => (p or q)'], ints = False)
truthtable.as_pandas().style.set_properties(**{'text-align': 'center'}).hide_index()
```
